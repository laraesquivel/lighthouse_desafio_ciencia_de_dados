Este notebook descreve o processo de modelagem de um modelo regressivo de previsão de preços, regressivo pois os preços são númericos.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import KFold
from sklearn.metrics import r2_score
from sklearn.ensemble import RandomForestRegressor
import matplotlib.pyplot as plt
import seaborn as sns
```

## Transformações 

```python
df = pd.read_csv('teste_indicium_precificacao.csv')
df.head()
```

Como vimos na análise exploratória temos registros de imóveis com valor zero, eles vão ser removidos pois eu considerei como ruído ter dados de uma plataforma de imóveis com valor 0. Outra coisa é que eu vou remover de incio o nome do host e a data da ultima_review. Criarei outras feature host_price_mean dado o host_id, que é a média de preços dos imoveis de cada host e preencherei as reviews por mês com o número medio de reviews no bairro.

```python
df['host_price_mean'] = df.groupby('host_id')['price'].transform('mean')
df.head()
```

```python
df = df[df['price'] != 0].drop(['host_name','ultima_review'], axis=1)
```

```python
media_review = df.groupby('bairro')['reviews_por_mes'].transform('mean')
df['reviews_por_mes'] = df['reviews_por_mes'].fillna(media_review)
df.head()
```

O bairro e o grupo de bairro tem pesos diferentes quando se for precificar cada um. Não sei examente o quanto cada um de fato interfere para usar o LabelEnconder, todavia não é um caso para o OneHotEnconder. Por isso também vou criar mas duas features bairro_price e bairro_group_price que se baseia na media de preços no bairro e no grupo de bairro.

```python
df['bairro_group_price'] = df.groupby('bairro_group')['price'].transform('mean')
df['bairro_price'] = df.groupby('bairro')['price'].transform('mean')
df.head()
```

Para room_type vamos codificar por label cada uma das categorias

```python
kinds_room = {'Shared room' : 0,  'Private room' : 1, 'Entire home/apt':2}
df['room_type'] = df['room_type'].map(kinds_room)
df.head()
```

Vou verificar se ainda resta algum vazio na base de dados, achei estranho a presença de três vazios em reviews por mês, ocorreu pois eles são os únicos representantes do seu bairro e por isso vou substituir pela média do grupo de bairro

```python
df.isna().sum()
```

```python
df[df['reviews_por_mes'].isna()]
```

```python
media_review = df.groupby('bairro_group')['reviews_por_mes'].transform('mean')
df['reviews_por_mes'] = df['reviews_por_mes'].fillna(media_review)
df.head()
```

Agora vou usar inicialmente somente os atributos númericos dos dados, tentei posteriormente incluir bairros e grupos de bairros com One Hot Enconder todavia não houve efeitos nas métricas. O modelo escolhido foi o Random Forest Regression, é um modelo para prever dados em contextos não lineares a partir da média dos valores previstos por n árvores de decisão. Não utilizei árvore de decisão porque os dados não parecem tão simples, o SVR depende muito de escala. O Random Forest foi escolhido porque como eu utilizei a média de bairros e de grupo de bairros como features, tive medo de ocorrer um sobreajuste e a floresta tende a evitar esse comportamento, além de lidar com dados desbalanceados diferente por exemplo do SVR. O maior problema é que ela não é interpretável como se fosse um KNN.

A métrica de precisão escolhida foi o Coeficiente de Determinação, ele mede a proporção da variância nos dados dependentes que é explicada pelo modelo, quanto mais próximo de 1 indica o quão bem o modelo explica os dados. Eu fiquei em dúvida entre o MSE, o Coeficiente e o MAE. Escolhi pois como temos dados muito discrepantes o MSE vai penalizar muito os valores maiores, o MAE é mais robusto aos outliers e não vai penalizar grandes erros como MSE, todavia o Coeficiente de  Determinação ele é mais légivel e compreemsivel de saber o quão bem o modelo estar pois vai somente de 0 a 1.

```python
df.columns
```

```python
df_cutted = df[['latitude',
       'longitude', 'room_type', 'price', 'minimo_noites', 'numero_de_reviews',
       'reviews_por_mes', 'calculado_host_listings_count',
       'disponibilidade_365', 'host_price_mean', 'bairro_group_price',
       'bairro_price']]
df_cutted.head()
```

```python
k = 5
kf = KFold(n_splits=k, shuffle=True, random_state=42)
```

```python
X = df[['latitude', 'longitude', 'room_type', 'minimo_noites', 'numero_de_reviews', 
        'reviews_por_mes', 'calculado_host_listings_count', 'disponibilidade_365', 
        'host_price_mean', 'bairro_group_price', 'bairro_price']]
y = df['price']
```

```python
X.isna().sum()
```

```python
random_forest = RandomForestRegressor(random_state=42, n_estimators=200)
```

```python
mae_scores = []
mse_scores = []
y_tests = []
y_preds = []

# Executando a validação cruzada
for train_index, test_index in kf.split(X):
    # Divida os dados em treino e teste com base nos índices dos folds
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    
    # Treinando o modelo
    random_forest.fit(X_train, y_train)
    
    # Predições
    y_pred = random_forest.predict(X_test)
    
    y_tests.append(y_test)
    y_preds.append(y_pred)


```

```python
r2_scores = []
for t,p in zip(y_tests, y_preds):
    r2 = r2_score(t,p)
    print(f'R2:{r2}')
    r2_scores.append(r2)

print(f'R2 medio:{sum(r2_scores)/k}')
```

O número de estimadores foi escolhido em cima de testagem, verifiquei também alterando atributos, essa estrutura foi a que obteve o melhor coeficiente médio, todavia não o melhor resultado entre todos os folds. Ao final podemos perceber que o modelo se saiu muito bem no primeiro, terceiro e quarto fold, todavia se saiu mal do segundo e no último fold. Ao olhar o gráfico de residuos observamos que temos pontos que o modelo conseguiu representar bem, ao mesmo que tempo que reve um erro residual maior.

```python
residuos = y_test - y_pred

# Gráfico de resíduos
plt.figure(figsize=(6, 5))
plt.scatter(y_test, residuos, color='red', alpha=0.7)
plt.axhline(y=0, color='black', linestyle='--', linewidth=1)
plt.xlabel("Valores reais")
plt.ylabel("Resíduo (Real - Previsto)")
plt.title("Gráfico de Resíduos")
plt.show()
```
